**2. Which genre of game has been released the most in a single year?**

```python
year_max_df = df.groupby(['Year', 'Genre']).size().reset_index(name='count')
year_max_idx = year_max_df.groupby(['Year'])['count'].transform(max) == year_max_df['count']
year_max_genre = year_max_df[year_max_idx].reset_index(drop=True)
year_max_genre = year_max_genre.drop_duplicates(subset=["Year", "count"], keep='last').reset_index(drop=True)
```

```python
genre = year_max_genre['Genre'].values
```

```python
plt.figure(figsize=(30, 15))
g = sns.barplot(x='Year', y='count', data=year_max_genre)
index = 0
for value in year_max_genre['count'].values:

    g.text(index, value + 5, str(genre[index] + '----' +str(value)), color='#002', size=14, rotation= 90, ha="center")
    index += 1

plt.xticks(rotation=90)
plt.show()
```

**3. How show the sales contribution of each category or genre?**

```python
labels = df["Genre"].value_counts().index
sizes = df["Genre"].value_counts()
colors = ['red','blue','brown','orange',"yellow","pink"]
plt.figure(figsize = (16,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors, startangle=45)
plt.title('Distribution of games by genre through pie chart',color = 'green',fontsize = 17)

plt.show()   
```

**4. How to show the relation between different columns through various plots?**

```python
# Creating pair for different columns to show relation between them    
    
df_pair = df.loc[:,["Year","Platform", "Genre", "NA_Sales","EU_Sales", "Other_Sales"]]
df_pair

sns.pairplot(df_pair, hue='Genre')
```

**5. Show total sales by region / area through chart. Also show which region had the highest sale?**

```python
top_sale_reg = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
top_sale_reg = top_sale_reg.sum().reset_index()
top_sale_reg = top_sale_reg.rename(columns={"index": "region", 0: "sale"})
top_sale_reg

plt.figure(figsize=(12, 6))
sns.barplot(x='region', y='sale', data = top_sale_reg)
```
